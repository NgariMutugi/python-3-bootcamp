```python
# add a second argument to the function, 'w' which stands for write
# passing 'w+' lets us read and write to the file
my_file = open('test.file', 'w+')
```

```python
# write the file
my_file.write('this is a new file ')

# seek the file to index 0

my_file.seek(0)

# read the file
my_file.read()

#close the file
my_file.close()
```
